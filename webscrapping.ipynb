```python
import requests
from bs4 import BeautifulSoup
r = requests.get("https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG")
c = r.content
c
```

```python
soup = BeautifulSoup(c,"html.parser")
print(soup.prettify())
```

```python
all = soup.find_all("div",{"class":"_2kHMtA"})
all
```

```python
len(all)
```

```python
all[0].find_all("div",{"class":"_4rR01T"})[0].text
```

```python
for item in all:
    print(item.find_all("div",{"class":"_4rR01T"})[0].text)
    print(item.find_all("div",{"class":"_30jeq3 _1_WHN1"})[0].text)
    print(item.find_all("li",{"class":"rgWa7D"})[0].text)
    print("\n")
```

```python
l =[]
for item in all:
    d = {}
    d["Mobile Name"] = item.find_all("div",{"class":"_4rR01T"})[0].text
    d["Rate"] = item.find_all("div",{"class":"_30jeq3 _1_WHN1"})[0].text.replace("","Rs.")
    d["Capacity"] = item.find_all("li",{"class":"rgWa7D"})[0].text
    l.append(d)
l
```

```python
import pandas
df = pandas.DataFrame(l)
df
df.to_csv("Samsung Mobiles.csv")
```
